```python
import pandas as pd
import numpy as np
```

```python
amazon = pd.read_csv('/Users/ale_d/Desktop/MLops_Projects/Datasets_Streaming_limpios/amazon.csv')
disney = pd.read_csv('/Users/ale_d/Desktop/MLops_Projects/Datasets_Streaming_limpios/disney.csv')
hulu = pd.read_csv('/Users/ale_d/Desktop/MLops_Projects/Datasets_Streaming_limpios/hulu.csv')
netflix = pd.read_csv('/Users/ale_d/Desktop/MLops_Projects/Datasets_Streaming_limpios/netflix.csv')
```

```python
# AÃ±adir columna de plataforma en cada dataset
netflix['plataforma'] = 'netflix'
amazon['plataforma'] = 'amazon'
hulu['plataforma'] = 'hulu'
disney['plataforma'] = 'disney'
```

```python
# Unir los datasets
df = pd.concat([netflix, amazon, hulu, disney])
```

```python
non_finite = np.isfinite(df["duration_int"])
df = df[non_finite]
df["duration_int"] = df["duration_int"].astype(int)
df["duration_int"] = df["duration_int"].fillna(-1).astype(int)
```

```python
rating_dict = {'tv-ma': 27, 'tv-14': 26,'r': 25,'13+': 24,'tv-pg': 23,'16+': 22,'all': 21,'18+': 20,'g': 19,'pg-13': 18,
               'pg': 17,'tv-g': 16,'tv-y7': 15,'tv-y': 14,'7+': 13,'nr': 12,'tv-nr': 11,'unrated': 10,'tv-y7-fv': 9,
               'nc-17': 8,'ages_18_': 7,'not_rate': 6,'ur': 5,'ages_16_': 4,'all_ages': 3,'16': 2,'not rated': 1}
```

```python
df['rating_num'] = df['rating'].map(rating_dict).astype(float)
```

```python
df.drop(['i'],axis=1, inplace=True)
```
